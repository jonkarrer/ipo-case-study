## 2. Perform summary statistics on the groups.

### Group 1: Banks/Investment Firms

#### Simple Pandas Summary

```python
bank_columns = [
    'citigroup', 'citi', 'morgan stanley', 'ubs', 'barclays', 'wells fargo', 
    'goldman sachs', 'deutsche bank', 'credit suisse', 'merrill lynch', 
    'rbc capital', 'jefferies', 'stifel', 'morgan stanley.1', 'merrill lynch.1', 
    'goldman sachs.1', 'raymond james', 'piper jaffray', 'robert w. baird', 
    'william blair', 'william blair.1', 'hill road'
]
print("*** BANKS *** \n", original_df[bank_columns].info())
print("*** BANKS *** \n", original_df[bank_columns].describe())
```

#### Distribution Plots

```python
cols = bank_columns
n = len(cols)

n_rows = 6
n_cols = 3

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 24))
axes = axes.flatten()

for ax, col in zip(axes, cols):
    sns.histplot(data=original_df, x=col, kde=True, ax=ax)
    ax.set_title(col)

for ax in axes[n:]:
    ax.set_visible(False)

plt.tight_layout()
plt.show()
```

```python
cols = bank_columns
n = len(cols)

n_cols = 4
n_rows = 5

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 24))
axes = axes.flatten()

for ax, col in zip(axes, cols):
    sns.scatterplot(data=original_df, x=col, y='diff', ax=ax)
    ax.set_title(col)

for ax in axes[n:]:
    ax.set_visible(False)

plt.tight_layout()
plt.show()
```
